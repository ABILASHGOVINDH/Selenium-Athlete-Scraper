```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Initialize WebDriver
driver = webdriver.Chrome()
driver.get("https://worldathletics.org/records/by-category/world-records")
driver.implicitly_wait(10)

# Accept cookies if the prompt appears
try:
    accept = driver.find_element(By.XPATH, "//button[text()='Allow all cookies']")
    accept.click()
    print("Cookies accepted")
except:
    print("No cookies prompt found")

# Click on the "Mixed" tab
try:
    mixed_tab = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//button[text()="mixed"]'))
    )
    mixed_tab.click()
    print("Mixed tab clicked.")
except Exception as e:
    print(f"Error clicking on 'Mixed' tab: {e}")
    driver.quit()

# Click on "United States" link
try:
    state = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//a[text()='United States']"))
    )
    state.click()
    print("United States link clicked.")
except Exception as e:
    print(f"Error clicking on 'United States': {e}")
    driver.quit()

# Competitors list
competitors = ["Vernon NORWOOD", "Shamier LITTLE", "Bryce DEADMON", "Kaylyn BROWN"]

# Initialize an empty DataFrame for progression data
all_progression_data = pd.DataFrame()

for competitor in competitors:
    try:
        # Click on the competitor's link
        competitor_element = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, f"//a[text()='{competitor}']"))
        )
        competitor_element.click()
        print(f"Processing {competitor}...")

        # Load all performances
        while True:
            try:
                see_more = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.XPATH, "//button/span[text()='SEE MORE PERFORMANCES']"))
                )
                see_more.click()
                print(f"Loaded more performances for {competitor}")
                time.sleep(1)  # Allow data to load
            except Exception:
                print("No more 'See More' button or unable to click.")
                break

        # Load progression data
        try:
            progression_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//button[text()='Progression']"))
            )
            progression_button.click()
            print("Progression data loaded.")

            # Extract progression table data
            data = []
            event_tables = driver.find_elements(By.CLASS_NAME, "profileStatistics_table__1o71p")
            for event_table in event_tables:
                rows = event_table.find_elements(By.CSS_SELECTOR, "tbody.profileStatistics_tableBody__1w5O9 tr")
                
                # Extract event name (this might change depending on the page structure)
                event_name = driver.find_element(By.CLASS_NAME, "profileStatistics_tableName__2qDVZ").text.strip()

                for row in rows:
                    cells = row.find_elements(By.TAG_NAME, 'td')
                    if cells:
                        year = cells[0].text.strip()
                        performance = cells[1].text.strip()
                        location = cells[2].text.strip()
                        date = cells[3].text.strip()

                        # Check if the performance is a world record
                        is_world_record = "world record" in performance.lower()  # Customize this condition if needed

                        # Append to data list
                        data.append([competitor, event_name, performance, year, location, date, is_world_record])

            # Add competitor progression data to DataFrame
            competitor_progression = pd.DataFrame(data, columns=["Name", "Event Name", "Performance", "Year", "Location", "Date", "Is World Record"])
            all_progression_data = pd.concat([all_progression_data, competitor_progression], ignore_index=True)
            print(f"Progression data for {competitor} extracted.")

        except Exception as e:
            print(f"Error loading progression data for {competitor}: {e}")

        # Navigate back to the main page
        driver.back()

    except Exception as e:
        print(f"Error processing {competitor}: {e}")
        driver.back()

# Save progression data to a CSV file
all_progression_data.to_csv("all_progression_data_mixed.csv", index=False)
print("All progression data saved to 'all_progression_data_mixed.csv'.")

# Close the browser
driver.quit()
```
